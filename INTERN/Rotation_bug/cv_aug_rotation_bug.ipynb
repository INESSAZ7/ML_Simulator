## Step-1:
### Код с ошибкой

```python
import cv2
import numpy as np


def rotated_image(image: np.ndarray, angle: int = 45) -> np.ndarray:
    """Rotate image by angle degrees."""
    width, height, channels = image.shape
    transform = cv2.getRotationMatrix2D((height / 2, width / 2), angle, 1)
    result = cv2.warpAffine(image, transform, (width, height))
    return result
```

```python
from skimage import data 
import matplotlib.pyplot as plt


pic = data.cat()
pic_transform = rotated_image(pic, 180)
fig, axes = plt.subplots(1,2,figsize=(7, 7))
axes[0].imshow(pic)
axes[1].imshow(pic_transform)
axes[0].set_title("Original")
axes[1].set_title("Rotated")

plt.show()
```

### Тест

```python
def test_rotated_image():
    image = data.cat()
    result = rotated_image(image)
    assert image.shape == result.shape
```

```python
test_rotated_image()
```

**NOTE:**
* В numpy нотации размерность картинки задается как (ширина, высота, канал)
* OpenCV – (высота, ширина)
